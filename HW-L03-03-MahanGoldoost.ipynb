# Exercise 2

```python
df_Hr['Position']
```

```python
df_Hr['PositionID']
```

```python
grouped_position = df_Hr.groupby('Position')['Absences'].sum().reset_index()
grouped_position.index = grouped_position.index + 1
grouped_position = grouped_position.sort_values(by='Absences', ascending=False)
grouped_position
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x='Absences', y='Position', data=grouped_position, hue='Position', palette='viridis', legend=False)
plt.title('Absences of each Positions')
plt.show()
```

As you can see in the barplot most absences are for Production Technician I and Production Technician II and Area Sales Manager

# Exercise 3

```python
df_Hr['IsTerminated'] = df_Hr['DateofTermination'].notna()
df_Hr['IsTerminated']
```

```python
plt.Figure(figsize=(8,6))
sns.boxplot(data=df_Hr,x='IsTerminated',y='EmpSatisfaction',hue='IsTerminated',palette={False : 'green',True : 'yellow'},legend=False)
plt.xticks([False,True],['Working','NotWorking'])
plt.xlabel('Work Status')
plt.title('EmpSatisfaction between epms that are working or not')
plt.tight_layout()
plt.show()
```

As you can see there is not much diffrenece between Emps that left the company and the ones that still working

# Exercise 4

```python
df_Hr['YearOfHire'] = pd.to_datetime(df_Hr['DateofHire']).dt.year
df_Hr['YearOfHire']
```

```python
grouped_Hired = df_Hr.groupby('YearOfHire')['EmpID'].count().reset_index()
grouped_Hired
```

```python
grouped_Hired.index = grouped_Hired.index + 1
grouped_Hired = grouped_Hired.sort_values(by='EmpID',ascending=False)
grouped_Hired
```

```python
plt.Figure(figsize=(12,8))
sns.barplot(data=grouped_Hired,x='YearOfHire',y='EmpID',hue='YearOfHire',palette='tab10',legend=False)
plt.title('Hired Emps Per Year')
plt.xlabel('Hired Year')
plt.ylabel('Count Of Hired Emps')
plt.tight_layout()
plt.show()
```

The most emps hired are for years: 2011 and 2014

# Exercise 5

```python
df_Hr[['SpecialProjectsCount','Absences']]
```

```python
plt.Figure(figsize=(7,5))
sns.scatterplot(data=df_Hr,x='SpecialProjectsCount',y='Absences',hue='SpecialProjectsCount',palette='viridis',legend=False)
plt.title('Relation Between Special Projects And Absences')
plt.xlabel('Special Projects')
plt.ylabel('Absences')
plt.show()
```

```python
corr_SpA = df_Hr[['SpecialProjectsCount','Absences']].corr()
corr_SpA
```

```python
plt.Figure(figsize=(6,4))
sns.heatmap(data=corr_SpA, annot=True, cmap='crest', fmt=".2f", vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

There is almost no relationship between special projects count and absences
